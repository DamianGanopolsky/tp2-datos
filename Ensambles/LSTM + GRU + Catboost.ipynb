```
import sys
sys.path.insert(1, '../FeatureEngineering')

import pandas as pd
import numpy as np
import tensorflow as tf
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from keras.preprocessing import sequence
from tensorflow import keras
from tensorflow.keras import layers
from gensim.models import KeyedVectors
from keras.preprocessing.text import Tokenizer
from keras.models import Sequential
from keras.layers import Dense, Conv1D, Flatten, MaxPooling1D, Embedding, GlobalMaxPooling1D, Dropout, LSTM,Input,Activation,Bidirectional,GRU
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import nltk
from sklearn.metrics import f1_score
import feature_builder
import util
import hyperparameter_tuning
embeddings = None
from hyperopt import hp,STATUS_OK
from catboost import CatBoostClassifier
```

```
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
Y = train_df['target'].values
```

```
def prepare_for_nn(df):
    processed = feature_builder.process_dataset(df,encoding_type='mean')
    return (processed, processed.to_numpy().reshape(processed.shape[0], 1, processed.shape[1]))
```

```
train_processed_catboost = feature_builder.process_dataset(train_df,encoding_type='mean',use_spacy=False)
train_processed, X_train_processed = prepare_for_nn(train_df)
```

```
ids = test_df['id']
test_processed_catboost = feature_builder.process_dataset(test_df,encoding_type='mean',use_spacy=False)
test_processed, X_test_processed = prepare_for_nn(test_df)
```

```
catboost = CatBoostClassifier()
catboost.fit(train_processed_catboost, train_df['target'])
```

```
y_train = train_df['target'].values
```

```
def LSTM():
    model = tf.keras.Sequential([
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(512),input_shape=(X_test_processed.shape[1], X_test_processed.shape[2])),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dropout(0.6924965129234393),
    tf.keras.layers.Dense(1)])
    return model
```

```
model_lstm = LSTM()
model_lstm.summary()
model_lstm.compile(loss=keras.losses.BinaryCrossentropy(from_logits=True),
              optimizer=keras.optimizers.Adam(1e-4),
              metrics=['accuracy'])
```

```
history_lstm = model_lstm.fit(X_train_processed, y_train, epochs=50, shuffle=False, validation_split=0.0)
```

```
def GRU():
    model = tf.keras.Sequential([
    tf.keras.layers.Bidirectional(tf.keras.layers.GRU(512),input_shape=(X_test_processed.shape[1], X_test_processed.shape[2])),
    tf.keras.layers.Dense(32 , activation='relu'),
    tf.keras.layers.Dropout(0.4961765821337532),
    tf.keras.layers.Dense(1)])
    return model
```

```
model_gru = GRU()
model_gru.summary()
model_gru.compile(loss=keras.losses.BinaryCrossentropy(from_logits=True),
              optimizer=keras.optimizers.Adam(1e-4),
              metrics=['accuracy'])
```

```
history_gru = model_gru.fit(X_train_processed, y_train, epochs=50, shuffle=False, validation_split=0.0)
```

```
catboost_prediction = catboost.predict(test_processed_catboost)
lstm_prediction = model_lstm.predict_classes(X_test_processed)
gru_prediction = model_gru.predict_classes(X_test_processed)
```

```
catboost_prediction = catboost_prediction.reshape(catboost_prediction.shape[0],1).astype('int32')
```

```
final_value = np.zeros(shape=(catboost_prediction.shape[0])).astype('int32')
for i in range(len(catboost_prediction)):
  value = catboost_prediction[i] + lstm_prediction[i] + gru_prediction[i]
  if value > 1:
    final_value[i] = 1
  else:
    final_value[i] = 0

final_value
```

```
final_df = pd.DataFrame({'target': final_value}, index=ids)
final_df
```
